```python
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
import pandas as pd
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#Normalizing the values
X_test = X_test/255
X_train = X_train/255
```

```python
#Flattening the dataset using reshape function from pandas
X_train_flatten = X_train.reshape(len(X_train), 28*28)
```

```python
X_train_flatten.shape
```

```python
X_test_flatten = X_test.reshape(len(X_test), 28 * 28)
```

```python
X_test_flatten.shape
```

```python
X_test_flatten[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation = 'sigmoid')
])
model.compile(
    optimizer= 'adam',
    loss = 'sparse_categorical_crossentropy', 
    metrics = ['accuracy']
)
model.fit(X_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
plt.matshow
```

```python
y_predicted = model.predict(X_test_flatten)
y_predicted
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(x) for x in y_predicted]
y_predicted_labels[:7]
```

```python
y_test[:7]
```

```python
conf_mat = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
conf_mat
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(conf_mat, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, activation = 'sigmoid'),
    keras.layers.Dense(10, input_shape=(784,), activation = 'sigmoid')
])
model.compile(
    optimizer= 'adam',
    loss = 'sparse_categorical_crossentropy', 
    metrics = ['accuracy']
)
model.fit(X_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(conf_mat, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation = 'tanh'),
    keras.layers.Dense(10, activation = 'softmax')
])
model.compile(
    optimizer= 'adam',
    loss = 'sparse_categorical_crossentropy', 
    metrics = ['accuracy']
)
model.fit(X_train, y_train, epochs=5)
```
